```python
clustered_b2uk = pd.read_csv('clustered_b2uk.csv')
```

```python
onehot = pd.get_dummies(clustered_b2uk[['Categories']],prefix='', prefix_sep='')
onehot['Neighborhood_label']=clustered_b2uk['Neighborhood_label']

city_list=clustered_b2uk['City'].tolist()
label_list=clustered_b2uk['Neighborhood_label'].tolist()

def firstword(str):
  return re.split(',',str)[0]  

clusterToCity={}

for i in range(0, len(city_list)):
    clusterToCity[label_list[i]] = firstword(city_list[i]) 


#city_list=list(map(firstword, city_list))
#onehot['City_name']=city_list

fixed_columns = [onehot.columns[-1]] + list(onehot.columns[:-1])
onehot = onehot[fixed_columns]

clusterToCity

```

```python
neighborhood_features=onehot.groupby('Neighborhood_label').mean().reset_index()


city_name_list = neighborhood_features['Neighborhood_label'].tolist()

for i in range(0,len(city_name_list)):
    city_name_list[i] = clusterToCity[city_name_list[i]]
    
neighborhood_features['City_name'] = city_name_list


neighborhood_features.to_csv('neighborhood_features_b2uk.csv',index=False, header=False)
neighborhood_features['City_name'].to_csv('city_names_b2uk.csv',index=False, header=False)
neighborhood_features.drop(columns=['Neighborhood_label','City_name'], inplace=True)
neighborhood_features.to_csv('neighborhood_features_unlabelled_b2uk.csv',index=False, header=False)
```
